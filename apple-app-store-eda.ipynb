```python
def categorize_price(price):
    if price == 0:
        return 'Free'
    elif price < 5:
        return 'Cheap'
    elif price < 20:
        return 'Moderate'
    else:
        return 'Expensive'

df['Price_Band'] = df['Price'].apply(categorize_price)
```

---
`Observation` :
- Review column is numeric column that holds information about the number fo reviews users left while dowloading /using the app

- We have a long sequence of data when it comes to Reviews column :

min 0 max 22685334

- We can club this long sequence into a compressed form and presentation. In another column called ReviewCategory

- if an app has reviews between 0 and 10000 --> Less than 10K
- if an app has reviews between 10000 and 500000 --> Between 10K and 500K
- if an app has reviews between 500000 and 1000000 --> Between 500K and 1Mil
- if an app has reviews between 1000000 and 22685334 --> Million Plus

```python
df.Reviews.describe()
```

```python
df['ReviewCategory'] = 'NoReviewsProvided'
df.loc[(df['Reviews'] > 0) & (df['Reviews'] <= 10000.0),'ReviewCategory'] = 'Less than 10K'
df.loc[(df['Reviews'] > 10000) & (df['Reviews'] <= 500000.0),'ReviewCategory'] = 'Between 10K and 500K'
df.loc[(df['Reviews'] > 500000) & (df['Reviews'] <= 1000000),'ReviewCategory'] = 'Between 500K and 1Mil'
df.loc[(df['Reviews'] > 1000000) & (df['Reviews'] <= 22685334),'ReviewCategory'] = 'Million Plus'
df['ReviewCategory'].value_counts()
```

# `1` Analysis of App Release Trends Over Time

```python
df['Released_Year'] = pd.to_datetime(df['Released']).dt.year
release_trend = df['Released_Year'].value_counts().sort_index()

plt.figure(figsize=(14, 6))
sns.lineplot(x=release_trend.index, y=release_trend.values, marker='o')
plt.title('Number of Apps Released Over the Years')
plt.xlabel('Year')
plt.ylabel('Number of Apps')
plt.grid(True)
plt.show()
```

 `Explanation`: 
 ***This line plot will highlight the growth or decline in the number of apps released per year, providing insights into the popularity of app development over time***.

`2` ***Correlation Analysis Between Price and User Rating***

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Price', y='Average_User_Rating', data=df)
plt.title('Correlation Between Price and User Rating')
plt.xlabel('Price')
plt.ylabel('User Rating')
plt.grid(True)
plt.show()
```

- `Explanation`: 
***This scatter plot will visualize whether there's any significant relationship between how much users pay for an app and how they rate it**

# `4` *Revenue Estimation Based on Price and Downloads*

```python
df['Estimated_Revenue'] = df['Price'] * df['Reviews']
revenue_by_genre = df.groupby('Primary_Genre')['Estimated_Revenue'].sum().sort_values(ascending=False)

plt.figure(figsize=(12, 6))
sns.barplot(x=revenue_by_genre.index, y=revenue_by_genre.values)
plt.xticks(rotation=90)
plt.title('Estimated Revenue by Genre')
plt.xlabel('Genre')
plt.ylabel('Estimated Revenue')
plt.grid(True)
plt.show()
```

# `Explanation`: 
***This bar plot will rank genres by estimated revenue, providing insight into which types of apps are the most financially successful***.

---
# `5`. Yearly Comparison of Average App Size

```python
avg_size_per_year = df.groupby('Released_Year')['Size_Mb'].mean()

plt.figure(figsize=(14, 6))
sns.lineplot(x=avg_size_per_year.index, y=avg_size_per_year.values, marker='o')
plt.title('Average App Size Over the Years')
plt.xlabel('Year')
plt.ylabel('Average Size (MB)')
plt.grid(True)
plt.show()
```

# `Explanation`:
***This line plot will reveal trends in the average app size over the years, indicating if modern apps tend to be larger or more data-intensive than those in the past***
